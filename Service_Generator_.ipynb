#read pdf file

```python

def load_pdf(path):
  """Loads text from a PDF file.

  Args:
    path: Path to the PDF file.

  Returns:
    A string containing the text content of the PDF file.
  """
  with open(path, 'rb') as file:
    reader = PyPDF2.PdfReader(file)
    text = ""
    for page_num in range(len(reader.pages)):
      page = reader.pages[page_num]
      text += page.extract_text()
    return text

pdf_text = load_pdf("/content/service_generation_topic.pdf")
print(pdf_text)
```

```python
def extract_text_from_pdf(pdf_document):
    """
    Extracts text from a PDF file.

    Parameters:
        pdf_document (str): The path to the PDF file.

    Returns:
        str: The extracted text from the PDF.
    """
    try:
        with open(pdf_document, 'rb') as pdf_file:
            reader = PyPDF2.PdfReader(pdf_file)
            text = ""

            for page_num in range(len(reader.pages)):
                page = reader.pages[page_num]
                text += page.extract_text() or ""

            return text

    except FileNotFoundError:
        print(f"Error: The file '{pdf_document}' was not found.")
        return ""
    except Exception as e:
        print(f"An error occurred: {e}")
        return ""
```

```python
# Function to split text into chunks
def split_text_into_chunks(text):
    splitter = RecursiveCharacterTextSplitter(
        chunk_size=10,
        chunk_overlap=2,
        separators=["."]
    )
    chunks = splitter.split_text(text)

    return chunks

# Example
pdf_document = "/content/service_generation_topic.pdf"
extracted_text = extract_text_from_pdf(pdf_document)

if extracted_text:
    chunks = split_text_into_chunks(extracted_text)

    print(f"Number of chunks: {len(chunks)}")

    for chunk in chunks:
        print(chunk)
        print("---\n")
```

```python
output_file = 'sentences.txt'
with open(output_file, 'w') as file:
    for sentence in chunks:
        file.write(sentence + '\n')

print(f"Sentences have been saved to {output_file}")
```
